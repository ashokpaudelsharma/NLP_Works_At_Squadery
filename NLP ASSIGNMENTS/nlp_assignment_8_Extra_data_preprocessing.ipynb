```python
#IMPORTING LIBRARIES FOR GENERAL NLP

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from array import *
import functools
```

```python
#DOWNLOADING STOPWORDS

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#IMPORTING PORTERSTEMMER FOR STEMMING

from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
#GETTING AVAILABLE DATASET  : fake.csv

dataset = pd.read_csv("fake.csv")
dataset.tail()
```

```python
dataset.columns
```

```python
dataset.shape
```

```python
#DROPPING NaN values to get a cleaned- up dataset

dataset.dropna(axis=0, inplace=True)
dataset.reset_index(inplace= True)
dataset.tail()
```

```python
ds_shape=dataset.shape
ds_shape
```

```python
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, ds_shape[0]):
    review = re.sub('[^a-zA-Z0-9]', ' ', dataset['text'][i]) #Removing the punctuation symbols
    review = review.lower() #Converting the textual information to all-lowercase text
    review = review.split()# Splitting the sentence/s into list/s
    ps = PorterStemmer() 
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))] #Removing english stopwords
    review = ' '.join(review) #Joining the list elements together to form a string
    corpus.append(review) #accumuating the strings to create the corpus
```
