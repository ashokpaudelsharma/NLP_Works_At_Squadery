1. Create a frequency distribution containing the given samples:

```python
text_samples= ['the quick brown fox jumped over another quick fox and thus the fox was successfully jumped over.',
                'the quick black dog walked over the quick black cat and the walking over was critically harmful and fatal.',
                'the quick brown dog walked over the lazy brown squirrel and the squirrel was scarred because of being walked over.']
```

```python
#fdist_sample = FreqDist(nltk.word_tokenize(text_samples[0]))
#fdist_sample = FreqDist(nltk.word_tokenize(text_samples[1]))
fdist_sample = FreqDist(nltk.word_tokenize(text_samples[2]))
```

```python
fig, ax= plt.subplots(figsize=(10,5))
fdist_sample.plot(20)
```

2. Increment the count for the sample:

```python
fdist_sample= []
sample_count=0
for s in text_samples:
    fdist_sample.append(FreqDist(nltk.word_tokenize(s)))
    sample_count+=1
```

```python
fig, ax= plt.subplots(figsize=(10,5))
fdist_sample[2].plot(20)
```

3. count of the number of times a given sample occurred:

```python
sample_word=input('Enter the sample word, you wish to count:')
print()
ind=int(input('Enter the index of the sample amongst the given text_samples:'))
print()
print("The count of the number of times the word occured in the sample: ")
fdist_sample[ind][sample_word]
```

4. Find the total number of samples:

```python
for ind in range(3):
    print("Total number of word_token samples for sample_text indexed "+ str(ind)+": ")
    print(fdist_sample[ind].N())
```

5. Find the n most common samples and their frequencies:

```python
n=int(input('Enter the value of n to get the most common samples:'))
print()


for ind in range(3):
    print("The n most common word_token samples and the respective frequencies for sample_text indexed "+ str(ind)+": ")
    print(fdist_sample[ind].most_common(n))
```

```python
#Iterate over the samples
i=0
for sample in fdist_sample:
    print("Sample number "+str(i)+" is : ")
    print(sample)
    i+=1
```

6. Find the sample with the greatest count:

```python
fdist_counts=[]
for ind in range(3):
    print("Total number of word_token samples for sample_text indexed "+ str(ind)+": ")
    print(fdist_sample[ind].N())
    fdist_counts.append(fdist_sample[ind].N())

#print("The counts of the various samples: ")
#print(fdist_counts)

print("The count of the sample with the greatest count is: ")
MAX=max(fdist_counts)
print(MAX)

print("The index of the sample with the greatest count is: ")
print(fdist_counts.index(MAX))
```
