21. Develop an NP chunker that converts POS-tagged text into a list of tuples, where each tuple consists of a verb followed by a sequence of noun phrases and prepositions, e.g. the little cat sat on the mat becomes ('sat', 'on', 'NP')...

```python
import nltk
```

```python
treebank=nltk.corpus.treebank
```

```python
# Grammar and the chunk_parser being defined
grammar = r"""
    NPH: {<DT>?<JJ.*>*<NN.*>+} # Noun phrase
    VIN: {<VB><IN><NPH>}    # Verb followed by prepositions followed by noun- phrase
    """
chunk_parser = nltk.RegexpParser(grammar)
```

```python
#Creating a tuples object
tuples = set()
```

```python
# Finding the  required clauses , i.e, VIN combinations
for sent in treebank.tagged_sents():
    tree = chunk_parser.parse(sent)
    #print(tree)
    
    for subtree in tree.subtrees():
        #print(subtree)
        if subtree.label() == 'VIN':
            #print(subtree)
            #print("subtree[0][0]:")
            #print(subtree[0][0])
            #print("subtree[1][0]:")
            #print(subtree[1][0])

            tuples.add((subtree[0][0],subtree[1][0], "NP"))
    
```
