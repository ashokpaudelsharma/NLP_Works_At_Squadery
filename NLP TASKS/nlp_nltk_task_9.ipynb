9. Write a program to find all words that occur at least three times in the Brown Corpus.

```python
import nltk
```

```python
#Importing the Brown Corpus
from nltk.corpus import brown
```

```python
#Fetching all the brown corpus words and saving the values in the variable "brown_corpus_words"

brown_corpus_words= brown.words()
```

```python
# Putting the condition for word frequency to access words occuring at least three times in the Brown Corpus

freq = nltk.FreqDist(brown_corpus_words)
words_at_least_three_times = [w for w in freq.keys() if freq[w] >= 3]
```

```python
#Demonstrating all words that occur at least three times in the Brown Corpus

print(words_at_least_three_times)
```
